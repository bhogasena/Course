**Matplotlib**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%reload_ext autoreload
%autoreload 2 
```

**Functional Plot**

```python
x = np.linspace(0,5,10)
y = x**2
plt.plot(x,y)
plt.title('Squares')
plt.xlabel('Numbers')
plt.ylabel('Square of Numbers')
```

**Multiple Plots**

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.xlabel('1x')


plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.xlabel('2x')

plt.subplot(2,2,3)
plt.plot(x,y,'r')
plt.xlabel('3x')

plt.subplot(2,2,4)
plt.plot(x,y,'m')
plt.xlabel('4x')
```

**Figure Objects**

```python
fig = plt.figure(figsize=(5,5),dpi=100)
axes_1 = fig.add_axes([0.1,0.1,0.5,0.5])
axes_1.set_xlabel('Numbers')
axes_1.set_ylabel('Squares of Numbers')
axes_1.set_title('Squares Plot')
axes_1.plot(x,y,label='x/x2')
axes_1.plot(y,x,label='x2/x')
axes_1.legend(loc=0)

axes_2 = fig.add_axes([0.8,0.1,0.7,0.5])
axes_2.set_xlabel('Numbers')
axes_2.set_ylabel('Squares of Numbers')
axes_2.set_title('Squares Plot')
axes_2.plot(x,y,'r',label='x/x2')
axes_2.legend(loc=0)
axes_2.set_title('Test Plot')

```

**Sub Plots**

```python
fig, axes = plt.subplots(figsize=(12,4),nrows=1,ncols=3)
plt.tight_layout()
def subplot(index,title,color):
  axes[index].set_title(title)
  axes[index].set_xlabel('Numbers')
  axes[index].set_ylabel('Squares of Numbers')
  axes[index].plot(x,y,color)
colors=['g','r','b']
for i in range(0,3):
  subplot(i,'Plot'+(str(i)),colors[i])

```

**Appearance**

```python
fig = plt.figure(figsize=(6,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y, color='navy', alpha=0.9, lw=2, ls='--',marker='o', markersize=7,markerfacecolor='g')
axes.set_xlim([0,3])
axes.set_ylim([0,9])
axes.grid(color='0.5',dashes=(1,1,1,5))
axes.set_facecolor('0.8')
```

**Save Visualizations to File**

```python
fig.savefig('plot.pdf')
```

**DataFrames**

```python
df_ics = pd.DataFrame.from_dict({'temp':[5,10,15,20,25,30,35,40,45,50,55,60,65,70],
                                 'sales':[1,2,3,4,5,6,7,10,12,20,30,40,50,60]})
```

```python
from matplotlib import markers
df_ics.sort_values(by='temp',ascending=False)
np_arr = df_ics.values
temp = np_arr[:,0]
sales = np_arr[:,1]

fig = plt.figure(figsize=(6,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(temp,sales,'g',marker='o',markerfacecolor='red',ls='--')
axes.set_xlabel('Temperature')
axes.set_ylabel('Sales')
axes.set_title('Sales Vs Temp')
axes.grid()
axes.annotate('Good Sales',xy=(70,60),xytext=(50,60),
              arrowprops=dict(facecolor='blue'))

plt.bar(x,y)
```
